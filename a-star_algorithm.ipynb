```python
import queue as q
def a_star(graph, start, end):
  queue= q.PriorityQueue()
  queue.put(((8,8),[start]))
  while not queue.empty():
    node= queue.get()
    current=node[1][-1]
    if end==current[0]:
      print('path found '+str(node[1])+ ', cost '+str(node[0][0]))
      break
    cost=node[0][0]
    distance=node[0][1]
    cost -= distance
    for child in graph[current]:
      path=node[1][:]
      path.append(child)
      queue.put((((cost + graph[current][child][0]), graph[current][child][1]), path))
```

```python
# node: (cost+h, h)
graph={
 'S':{'A':(12,9), 'B':(2,1), 'C':(8,3)},
 'A':{'E':(8,1), 'G1':(10,0)},
 'B':{'C':(5,3), 'F':(7,5)},
 'C':{'G':(11,0)},
 'D':{'B':(5,1), 'S':(14,8), 'G2':(5,0)},
 'E':{'G3':(2,0)},
 'G1':{},
 'G2':{},
 'G3':{'F':(5,5)},
 'F':{'D':(5,4)}
}
```

```python
a_star(graph, 'S', 'G')
```
